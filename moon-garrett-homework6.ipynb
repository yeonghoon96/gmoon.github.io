### Heat Map Code:

```python
# bqplot.ColorScale?
# bqplot.OrdinalScale?
# bqplot.LinearScale?
```

```python
mySelectedLabel = ipywidgets.Label()
```

```python
def on_selected(change):
     if len(change['owner'].selected) == 1:
        i, j = change['owner'].selected[0]
        #print(i, j)
        v = buildings_heatmap_data.iloc[i,j] # grabbing the underlying datavalue associated with this row/column
        mySelectedLabel.value = 'Total Sqft = ' + str(v)
    
        
# def on_selected(change):
#     print(change)

# [1] set scales:
col_sc = bqplot.ColorScale(scheme = 'Reds')
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

# [2] axis:
c_ax = bqplot.ColorAxis(scale = col_sc, 
                        label='Sqft',
                        orientation = 'vertical', 
                        side = 'right')
x_ax = bqplot.Axis(scale = x_sc, label='Gov Department - Agency Name',
                   grid_lines='solid')
y_ax = bqplot.Axis(scale = y_sc, 
                   orientation = 'vertical',
                  label='Congress Districts')

# (3) Marks: heatmap
heat_map = bqplot.GridHeatMap(color = buildings_heatmap_data,
                              scales = {'color': col_sc,
                                        'row': y_sc,
                                        'column': x_sc},
                              interactions = {'click': 'select'}, # make interactive on click of each box
                              anchor_style = {'fill':'Blue'}, # to make our selection blue
                              selected_style = {'opacity': 1.0}, # make 100% opaque if box is selected
                              unselected_style = {'opacity': 0.8}) # make a little see-through if not

# (4) Interaction(s):
heat_map.observe(on_selected, 'selected')

# (5) Paint heatmap canvas, don't display yet:
fig_heatmap = bqplot.Figure(marks = [heat_map], axes = [c_ax, y_ax, x_ax],
                            title="Agency's Total SQFT in given Congress Dist")
#fig_heatmap

# Dashboard:
myDashboard = ipywidgets.VBox([mySelectedLabel, fig_heatmap])
# myDashboard
```

- Any aesthetic choices you made (colors, layout, plot size, label size) and why, and what you'd like to experiment with if you had more time.

Answer: 
Choice the standard reds coloring for the heatmap with a blue selection color, however, I wanted to make sure the x-axis specified not just numbers, but the actual Agency Name, so I decided to put all the unique column names into a list datatype in the variable called `col_names`. This was then inserted into the `bqplot.GridHeatMap` parameter called `column`. Initially it was too close together, so we rotated the Heat Map's x-axis using the `tick_rotate=#` parameter. This made it a little better, but the originally set Heat Map opacity was too dark, so I ajdusted the `select_style` and `unselected_style` parameters to lower values. This helped increased the visibility of the Grid's x-axis Agency Name ticks. Also made the effort to use a `VBox` when displaying the linked grid heat map and line plot instead of using `HBox`. Because of the custom Pivot Table, the original heat map was rather large horizontally as well as the accompanying line plot. It made most sense visually to just stack them vertically.
